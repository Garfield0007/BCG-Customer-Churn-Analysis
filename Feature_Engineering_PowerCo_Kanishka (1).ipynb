Remove Unnecessary Columns

```python
cols_to_drop = [
    'id',
    'date_end',
    'date_modif_prod',
    'forecast_price_pow_off_peak',
    'forecast_discount_energy'  # optional
]
df.drop(cols_to_drop, axis=1, inplace=True)
```

Convert Date Columns to Datetime

```python
# Ensure date columns are in datetime format
df['date_activ'] = pd.to_datetime(df['date_activ'], errors='coerce')
df['date_renewal'] = pd.to_datetime(df['date_renewal'], errors='coerce')
```

 Feature Engineering (Create New Features)

📅 Date-Based Features

```python
# Duration of contract
df['contract_days'] = (df['date_renewal'] - df['date_activ']).dt.days

# Month and year from date
df['month_renewal'] = df['date_renewal'].dt.month
df['activation_year'] = df['date_activ'].dt.year
```

📊 Ratio-Based Features

```python
# Forecast accuracy (closer to 1 is better)
df['consumption_ratio'] = df['cons_12m'] / (df['forecast_cons_12m'] + 1e-5)

# Margin per kWh forecasted
df['margin_per_kWh'] = df['net_margin'] / (df['forecast_cons_12m']+ 1e-5)

# Gross to Net margin ratio
df['margin_ratio'] = df['margin_net_pow_ele'] / (df['margin_gross_pow_ele'] + 1e-5)  # avoid divide by zero
```

🚩 Flag-Based & Combined Features

```python
# Dual service flag (gas + electricity)
df['dual_service'] = ((df['has_gas'] == 1) & (df['cons_gas_12m'] > 0)).astype(int)

# High consumption customer
df['high_usage_flag'] = (df['cons_12m'] > df['cons_12m'].median()).astype(int)

# Long contract flag
df['long_contract'] = (df['contract_days'] > 365).astype(int)
```

💸 Price Variability Features

```python
# Mean yearly price volatility
df['avg_yearly_price_var'] = df[[
    'var_year_price_off_peak',
    'var_year_price_peak',
    'var_year_price_mid_peak'
]].mean(axis=1)

# 6-month price change difference
df['avg_6m_price_var'] = df[[
    'var_6m_price_off_peak',
    'var_6m_price_peak',
    'var_6m_price_mid_peak'
]].mean(axis=1)

# Recent price increase signal
df['recent_price_spike'] = (df['avg_6m_price_var'] > df['avg_yearly_price_var']).astype(int)
```
